```python
from sklearn.ensemble import RandomForestClassifier
# Prepare data for modeling
X = df.drop(columns=['FTR_Encoded', 'Date', 'FTHG', 'FTAG', 'HTHG', 'HTAG'], axis=1)
y = df['FTR_Encoded']

# Fit a Random Forest model
model = RandomForestClassifier(random_state=42)
model.fit(X, y)

# Feature Importance Plot
importance = model.feature_importances_
features = X.columns
plt.figure(figsize=(12, 6))
sns.barplot(x=importance, y=features, palette="coolwarm")
plt.title("Feature Importance for Predicting Match Outcome")
plt.show()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, shuffle=True, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
from tensorflow.keras.callbacks import EarlyStopping
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout

# Initialize the model
model = Sequential()

# Input layer and hidden layers with Dropout
model.add(Dense(64, activation='relu', input_shape=(X_train.shape[1],)))
model.add(Dropout(0.3))
model.add(Dense(64, activation='relu'))
model.add(Dropout(0.3))
model.add(Dense(32, activation='relu'))

# Output layer for multiclass classification with 3 classes
model.add(Dense(3, activation='softmax'))

# Compile the model
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# Add EarlyStopping to prevent overfitting
early_stopping = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)
```

```python
history = model.fit(
    X_train_scaled, y_train,
    validation_data=(X_test_scaled, y_test),
    epochs=50, batch_size=32,
    callbacks=[early_stopping],
    verbose=1
)
```

```python
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier, ExtraTreesClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier
from catboost import CatBoostClassifier
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, roc_auc_score, confusion_matrix, roc_curve, auc
import numpy as np
from sklearn.preprocessing import label_binarize

# # Scale the data
# scaler = StandardScaler()
# X_train_scaled = scaler.fit_transform(X_train)
# X_test_scaled = scaler.transform(X_test)

# Define models
models = {
    "Random Forest": RandomForestClassifier(random_state=42),
    "Gradient Boosting": GradientBoostingClassifier(random_state=42),
    "Neural Network": MLPClassifier(random_state=42, max_iter=200),
    "Logistic Regression": LogisticRegression(max_iter=200, random_state=42),
    "Linear Discriminant Analysis": LinearDiscriminantAnalysis(),
    # "SVM": SVC(probability=True, random_state=42),
    "k-Nearest Neighbors": KNeighborsClassifier(n_neighbors=5),
    "Gaussian Naive Bayes": GaussianNB(),
    "Extra-Tree Classifier": ExtraTreesClassifier(class_weight='balanced'),
    "XGBoost": XGBClassifier(use_label_encoder=False, eval_metric='mlogloss', random_state=42),
    "LightGBM": LGBMClassifier(random_state=42),
    "CatBoost": CatBoostClassifier(verbose=0, random_state=42)
}

# Binarize y_test for multi-class ROC
y_test_binarized = label_binarize(y_test, classes=np.unique(y_test))
n_classes = y_test_binarized.shape[1]

# Evaluate models and compute metrics + ROC curves
plt.figure(figsize=(12, 8))

for name, model in models.items():
    print(f"\nEvaluating {name}...")
    
    # Train the model
    model.fit(X_train_scaled, y_train)
    
    # Predict class labels
    y_pred = model.predict(X_test_scaled)
    
    # Predict probabilities
    y_prob = model.predict_proba(X_test_scaled) if hasattr(model, "predict_proba") else None

    # Compute metrics
    accuracy = accuracy_score(y_test, y_pred)
    conf_matrix = confusion_matrix(y_test, y_pred)
    print(f"Accuracy: {accuracy:.4f}")
    print(f"Confusion Matrix:\n{conf_matrix}")
    
    if y_prob is not None:
        # Compute ROC curve and AUC for the macro-average
        fpr = dict()
        tpr = dict()
        roc_auc = dict()

        for i in range(n_classes):
            fpr[i], tpr[i], _ = roc_curve(y_test_binarized[:, i], y_prob[:, i])
            roc_auc[i] = auc(fpr[i], tpr[i])

        # Compute the macro-average ROC curve
        all_fpr = np.unique(np.concatenate([fpr[i] for i in range(n_classes)]))
        mean_tpr = np.zeros_like(all_fpr)
        for i in range(n_classes):
            mean_tpr += np.interp(all_fpr, fpr[i], tpr[i])
        mean_tpr /= n_classes

        roc_auc_macro = auc(all_fpr, mean_tpr)

        # Plot the macro-average ROC curve
        plt.plot(all_fpr, mean_tpr, label=f"{name} (AUC = {roc_auc_macro:.2f})")
    else:
        print(f"{name} does not support predict_proba; skipping ROC.")

# Plot diagonal
plt.plot([0, 1], [0, 1], 'k--')

# Customize the plot
plt.title("Macro-Average ROC Curves for Multi-class Classification")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.legend(loc="center left", bbox_to_anchor=(1, 0.5))
plt.grid()
plt.tight_layout()
plt.show()
```
